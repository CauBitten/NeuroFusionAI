```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sys.path.append(os.path.abspath('..'))
```

```python
from src import train_model
```

```python
history, model = train_model(
    data_dir='../data/raw',
    batch_size=20,
    epochs=50
)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Treino')
plt.plot(history.history['val_accuracy'], label='Validao')
plt.title('Acurcia')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Treino')
plt.plot(history.history['val_loss'], label='Validao')
plt.title('Perda')
plt.legend()
plt.show()
```
